```python
%matplotlib inline

### Our First Machine Learning ###

print "import iris dataset from scikit learn"
from sklearn import datasets
iris = datasets.load_iris()
X_iris, y_iris = iris.data, iris.target
print X_iris.shape, y_iris.shape
print X_iris[0], y_iris[0]

print "\nTrain the data with 75% of data randomly sampled, only use sepal length and width, and normalize the data"
from sklearn.cross_validation import train_test_split
from sklearn import preprocessing
import numpy as np
X, y = X_iris[:, :2], y_iris
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=33)
print X_train.shape, y_train.shape
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print "plot the data as a scatter plot"
import matplotlib.pyplot as plt
colors = ['red', 'greenyellow', 'blue']
for i in xrange(len(colors)):
   xs = X_train[:,0][y_train == i]
   ys = X_train[:,1][y_train == i]
   plt.scatter(xs, ys, c=colors[i])
plt.legend(iris.target_names)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.draw()
plt.show()
```

```python
print "\nCreate a linear classifier and fit it with training data"
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier()
clf.fit(X_train, y_train)
print clf.coef_
print clf.intercept_
```

```python
print "draw all three plots"
x_min, x_max = X_train[:, 0].min() - .5, X_train[:, 0].max() + .5
y_min, y_max = X_train[:, 1].min() - .5, X_train[:, 1].max() + .5
xs = np.arange(x_min, x_max, 0.5)
fig, axes = plt.subplots(1, 3)
fig.set_size_inches(10, 6)
for i in [0, 1, 2]:
    axes[i].set_aspect('equal')
    axes[i].set_title('Class ' + str(i) + ' versus the rest')
    axes[i].set_xlabel('Sepal length')
    axes[i].set_ylabel('Sepal width')
    axes[i].set_xlim(x_min, x_max)
    axes[i].set_ylim(y_min, y_max)
    plt.sca(axes[i])
    plt.scatter(
        X_train[:, 0], 
        X_train[:, 1], 
        c=y_train, 
        cmap=plt.cm.prism
    )
    ys = (-clf.intercept_[i] - xs * clf.coef_[i, 0]) / clf.coef_[i, 1]
    plt.plot(xs, ys, hold=True)
plt.draw()
plt.show()
```

```python
print "predict the value of an unknown iris"
print clf.predict(scaler.transform([[4.7, 3.1]]))

print "show the classifier decision function"
print clf.decision_function(scaler.transform([[4.7, 3.1]]))
```
